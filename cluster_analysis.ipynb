# 1. Test 10 APK

## 1.1 Read 10 APK data for test

```python
%%time
%matplotlib inline

import os
import numpy as np
import joblib as jl

FEA_ROOT = 'E:\\project\\data_mining_research\\working\\fea_tmp'# 10 apk data for test

X_label = []
X = []
for f1 in os.listdir(FEA_ROOT)[0:10]:
    f1_full_path =  os.path.join(FEA_ROOT, f1)
    for f2 in os.listdir( f1_full_path ):
        f2_full_path = os.path.join(f1_full_path, f2)
        X_label.append('{}___{}'.format(f1, f2))
        X.append(jl.load(os.path.join(f2_full_path, 'fea.jl')))
X = np.array(X)
```

```python
X.shape # we get 2228 packages from 10 apk
```

## 1.2 cluster the 2228 package data

```python
%%time
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

random_state = 170
y_pred = KMeans(n_clusters=1000, random_state=random_state).fit_predict(X)
```

```python
stat = {}
for i in range(1000):
    stat[i]=0
for i in y_pred:
    stat[i] += 1
for k in stat:
    if(stat[k]>1):
        print('{}:{}'.format(k, stat[k]))
        for i in range(len(y_pred)):
            if(y_pred[i]==k):
                print('{}:{}'.format(k, X_label[i]))
```

## 1.3 distance calculation

```python
import math

def get_eu_dist(x, y):
    assert x.shape==y.shape
    s = 0
    for i in range(x.shape[0]):
        d = (x[i]-y[i])*(x[i]-y[i])
        s += d
    return math.sqrt(s)
```

```python
from tqdm import *

result = []
th = 0.005 # threshold
i=0
for m in tqdm(X): 
    if(i==0):
        d = {}
        d['count']=1
        d['name']=[]
        d['name'].append(X_label[i])
        d['index']=[]
        d['index'].append(i)
        result.append(d)
    else:
        x = X[i]
        found = False
        for j in range(len(result)):
            y = X[result[j]['index'][0]]
            dist = get_eu_dist(x,y)
            if(dist<th):
                result[j]['count'] += 1
                result[j]['name'].append(X_label[i])
                result[j]['index'].append(i)
                found = True
                break
        if(found==False):
            d = {}
            d['count']=1
            d['name']=[]
            d['name'].append(X_label[i])
            d['index']=[]
            d['index'].append(i)
            result.append(d)
    i += 1
```

```python
for v in result:
    if(v['count']>1):
        print(v)
```
